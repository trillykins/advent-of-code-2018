```python
import matplotlib.pylab as plt
from matplotlib.pyplot import figure
import os

w1 = open('wire1.txt', 'r').readlines()
w2 = open('wire2.txt', 'r').readlines()
wire1x = [(int(a.split(',')[0][3:]), int(a.split(',')[1][2:2+a.split(',')[1][2:].find('}')])) for a in open("wire1.txt", "r").readlines()]
wire2x = [(int(a.split(',')[0][3:]), int(a.split(',')[1][2:2+a.split(',')[1][2:].find('}')])) for a in open("wire2.txt", "r").readlines()]

wire1 = list()
wire2 = list()

for a, b in zip(w1, w2):
    wire1.append((int(a.split(',')[0][3:]), int(a.split(',')[1][2:2+a.split(',')[1][2:].find('}')])))
    wire2.append((int(b.split(',')[0][3:]), int(b.split(',')[1][2:2+b.split(',')[1][2:].find('}')])))
    
x1, y1 = zip(*wire1x)
x2, y2 = zip(*wire2x)

figure(num=None, figsize=(100, 100), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```
